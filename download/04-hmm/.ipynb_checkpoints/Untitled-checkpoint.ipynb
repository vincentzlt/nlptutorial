```python
import sys,pprint,math
from collections import defaultdict
```

```python
class pos_model():
    def __init__(self, f_name):
        self.pos_trans_dict = defaultdict(lambda: 0)
        self.pos_word_emi_dict = defaultdict(lambda: 0)
        self.pos_dict = defaultdict(lambda: 0)

        for line in open(f_name, "r", encoding="UTF-8"):
            word_pos_list = line.split(" ")
            #w_list=list()
            pos_list = list()
            for p in word_pos_list:
                self.pos_word_emi_dict[p] += 1
                w, pos_ = p.split("_")
                self.pos_dict[pos_] += 1
                #w_list.append(w)
                pos_list.append(pos_)
            for pos1, pos2 in zip(pos_list, pos_list[1:]):
                self.pos_trans_dict[pos1 + " " + pos2] += 1

    def calc_pos_trans_P(self, pos, pos_prev):
        P = self.pos_trans_dict[pos_prev + " " + pos] / float(
            self.pos_dict[pos_prev])
        return P

    def calc_pos_emission_P(self, w, pos):
        P = self.pos_word_emi_dict[w + "_" + pos] / float(self.pos_dict[pos])
        lambda_ = 0.95
        N = 1000000
        return lambda_ * P + (1 - lambda_) / float(N)
    
    def calc_line_tag_viterbi(self,line_str):
        line=line_str.split()
        len_line=len(line)
        tag_list=[]
        for w in self.pos_dict:
            tag_list.append(w)
                      
        line.append("</s>")
        line.insert(0,"<s>")
        
        best_score=[]
        best_path=[]
        
        
        for w in line:
            if w=="<s>":
                flag_start=True
            elif w=="</s>":
                pass
            else:
                if flag_start:
                    for t in tag_list:
                        
```
